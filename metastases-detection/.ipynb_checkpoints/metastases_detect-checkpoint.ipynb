```python
import numpy as np
import h5py
from torch.utils.data import Dataset
from PIL import Image

class MetastasesDataset(Dataset):
    """Metastases dataset."""

    def __init__(self, filename_x, filename_y, transform=None):
        """
        Args:
            filename_x: path of the filename of sample image
            filename_y: path of the filename of sample label
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        f_x = h5py.File(filename_x,'r')
        f_y = h5py.File(filename_y,'r')
        self.x = f_x['x']
        self.y = f_y['y']
        self.transform = transform
        assert len(self.x) == len(self.y)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        image = np.asarray(self.x[idx])
        image = Image.fromarray(image)
        label = np.squeeze(self.y[idx])

        if self.transform:
            image = self.transform(image)

        sample = (image, label)

        return sample
```

```python
##  Specify data loaders

### Write data loaders for training, validation, and test sets
## Specify appropriate transforms, and batch_sizes
import torch
import torchvision.transforms as transforms
from torch.utils.data.sampler import SubsetRandomSampler

use_cuda = torch.cuda.is_available()

train_transform = transforms.Compose([transforms.RandomResizedCrop(224), 
                                transforms.RandomHorizontalFlip(), 
                                transforms.RandomVerticalFlip(), 
                                transforms.RandomRotation(15),
                                transforms.ToTensor(),
                                transforms.Normalize((0.48,0.45,0.4),(0.229,0.224,0.225))])

data = {}
data['train'] = MetastasesDataset('/Users/fayehall/data/pcam/camelyonpatch_level_2_split_train_x.h5',
                           '/Users/fayehall/data/pcam/camelyonpatch_level_2_split_train_y.h5',
                           train_transform)
data['valid'] = MetastasesDataset('/Users/fayehall/data/pcam/camelyonpatch_level_2_split_valid_x.h5',
                           '/Users/fayehall/data/pcam/camelyonpatch_level_2_split_valid_y.h5',
                           train_transform)

data['test'] = MetastasesDataset('/Users/fayehall/data/pcam/camelyonpatch_level_2_split_test_x.h5',
                           '/Users/fayehall/data/pcam/camelyonpatch_level_2_split_test_y.h5',
                           train_transform)

for x in data:
    print(x, len(data[x]))
    
loaders = {}
batch_size = 20
loaders['train'] = torch.utils.data.DataLoader(data['train'], batch_size=batch_size, shuffle=True)
loaders['valid'] = torch.utils.data.DataLoader(data['valid'], batch_size=batch_size, shuffle=True)
loaders['test'] = torch.utils.data.DataLoader(data['test'], batch_size=batch_size, shuffle=True)


print(loaders['train'])
```

## Step 2: Model Architecture

```python
import torchvision.models as models
import torch.nn as nn

model = models.vgg16(pretrained=True)

for p in model.features.parameters():
    p.requires_grad = False

n_input = model.classifier[6].in_features
last_layer = nn.Linear(n_input, 1)
model.classifier[6] = last_layer

if use_cuda:
    model = model.cuda()
    
print(model)
```

## Step 3: Specify Loss Function and Optimizer

```python
import torch.optim as optim

criterion = nn.BCEWithLogitsLoss()

optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
```

## Step 4: Train and Validate the Model

```python
def train(n_epochs, loaders, model, optimizer, criterion, use_cuda, save_path):
    """returns trained model"""
    # initialize tracker for minimum validation loss
    valid_loss_min = np.Inf 
    
    for epoch in range(1, n_epochs+1):
        # initialize variables to monitor training and validation loss
        train_loss = 0.0
        valid_loss = 0.0
        
        ###################

        # train the model
        ###################
        model.train()
        for batch_idx, (data, target) in enumerate(loaders['train']):
            # move to GPU
            if use_cuda:
                data, target = data.cuda(), target.cuda()
            ## find the loss and update the model parameters accordingly
            ## record the average training loss, using something like
            ## train_loss = train_loss + ((1 / (batch_idx + 1)) * (loss.data - train_loss))
            optimizer.zero_grad()
            output = np.squeeze(model(data))
            target = target.type(torch.FloatTensor)
            
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            train_loss = train_loss + ((1 / (batch_idx + 1)) * (loss.data - train_loss))

            
        ######################    
        # validate the model #
        ######################
        model.eval()
        for batch_idx, (data, target) in enumerate(loaders['valid']):
            # move to GPU
            if use_cuda:
                data, target = data.cuda(), target.cuda()
            ## update the average validation loss
            output = model(data)
            loss = criterion(output, target)
            valid_loss = valid_loss + ((1 / (batch_idx + 1)) * (loss.data - valid_loss))


            
        # print training/validation statistics 
        print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(
            epoch, 
            train_loss,
            valid_loss
            ))
        
        ## TODO: save the model if validation loss has decreased
        if valid_loss <= valid_loss_min:
            torch.save(model.state_dict(), save_path)
            valid_loss_min = valid_loss
            print('Saving model...', valid_loss)

            
    # return trained model
    return model

n_epochs = 3
model = train(n_epochs, loaders, model, optimizer, criterion, use_cuda, 'model.pt')
```
